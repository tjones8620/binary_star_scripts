```python
""" Cross correlation """
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Generate a noisy signal to be correlated
np.random.seed(1234)
n = 1000
time = np.arange(n)
sig  = np.cos(0.3 * time) + np.random.normal(scale=0.3, size=time.shape)

# Correlate the signal with a reference signal
corr = signal.correlate(sig, np.ones(100), mode='same') / 100

# Plot the results
fig, (ax_orig, ax_noise, ax_corr) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.set_title('Original signal')
ax_orig.margins(0, 0.1)
ax_noise.plot(np.ones(100))
ax_noise.set_title('Reference signal')
ax_noise.margins(0, 0.1)
ax_corr.plot(corr)
ax_corr.set_title('Cross-correlated signal')
ax_corr.margins(0, 0.1)
fig.tight_layout()
plt.show()
```

```python
""" Perform gaussian fitting """

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the function to fit
def func(x, a, b, c):
    return a * np.exp(-(x - b)**2 / (2 * c**2))

# Generate some data to fit
xdata = np.linspace(0, 4, 50)
y = func(xdata, 2.5, 1.3, 0.5)
ydata = y + 0.2 * np.random.normal(size=len(xdata))

# Fit for the parameters a, b, c of the function `func`
popt, pcov = curve_fit(func, xdata, ydata)

# Plot the data along with the fitted curve
plt.plot(xdata, ydata, 'ko', label="Original Noised Data")
plt.plot(xdata, func(xdata, *popt), 'r-', label="Fitted Curve")
plt.legend()
plt.show()


```
